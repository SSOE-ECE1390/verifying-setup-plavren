```python
import cv2

import os
import numpy as np

import matplotlib.pyplot as plt
```

## Reading in the image

```python
#reading an image using openCV
apollo = cv2.imread(r'/Users/phillav/github-classroom/SSOE-ECE1390/verifying-setup-plavren/Data/Apollo_11_Launch.jpg')
```

```python
print("Image size (H, W) is:", apollo.shape)
print("Data type of image is:", apollo.dtype)
#reading and printing out the shape and data type of the 
#image (just for completeness)
```

## Displaying the image

```python
plt.imshow(apollo)
plt.title("Apollo 11 Launch in Grayscale")
#Using loaded image, displaying it (already in Grayscale, so no
#manipulation needed to do that)
```

## Cropping the middle part and displaying

```python
apolloMiddleCrop = apollo[0:600, 400:800] 
#crop pixels from 0-600 (vertical) and 400 to 800 (horizontal)
plt.imshow(apolloMiddleCrop)
plt.title("Apollo 11 Launch, Grayscale, Rocket Platform")
#use matplotlib to display again and add a title to describe what I cropped
```

## Taking crop, changing brightness

```python
matrix = np.ones(apolloMiddleCrop.shape, dtype="uint8") * 100

"""
introducing a matrix of ones, same shape/size as the image and uint8 (data type 
just like the image) and multiplying the ones by 100 to modulate brightness later
"""

apolloBright = cv2.add(apolloMiddleCrop, matrix)
#make img brighter by adding values

apolloDark = cv2.subtract(apolloMiddleCrop, matrix)
#make img darker by subtracting values

# Showing images
plt.figure(figsize=[10, 5])
plt.subplot(131); plt.imshow(apolloDark);         plt.title("Darker");
plt.subplot(132); plt.imshow(apolloMiddleCrop);   plt.title("Original");
plt.subplot(133); plt.imshow(apolloBright);       plt.title("Brighter");

#subplot works just like in Matlab!
#need to figure out how figsize works beyond just fiddling w it until it works
#can use semicolon ; to keep stuff on one line
```

## Rotating dark image

```python
apolloRotated = cv2.rotate(apolloDark, cv2.ROTATE_90_CLOCKWISE)
#rotating the image 90º clockwise using cv2

plt.imshow(apolloRotated)
plt.title("Apollo 11 Launch, Dark, Rotated")
#showing and titling the rotated image
```

## Saving the finally modified image

```python
#Saving file
cv2.imwrite("ApolloLaunch_Cropped_Dark_Rotated.jpg", apolloRotated)

#this makes the file save to the same folder you are working in
#in this case, it's 'verifying-setup-plavren'
```
